#AND OR Operators

```r
#and
a <- 100
b <- 22
c <- 300

if(a>b & c>a){
  print('Both conditions are ture')
}else{
  print('Not true')
}
cat('\n')

# OR
a <- 200
b <- 33
c <- 500

if (a > b | a > c) {
  print("At least one of the conditions is true")
}
```

#While Loop

```r
i <- 1
while(i < 6){
  print(i)
  i<-i +1
}
```

```r
#Break
i <- 1
while (i < 6) {
  print(i)
  i <- i + 1
  if (i == 4) {
    break
  }
}
```

```r
# next
i <- 0
while (i < 6) {
  i <- i + 1
  if (i == 3) {
    next
  }
  print(i)
}
```

#Data Structures


```r
# Vectors is used to store multiple values ​​of the same data type together.
numbers <- c(10,20,30,40,50,50)
numbers
```

```r
numbers[1]
numbers[3]
```

```r
#Lists are used to keep different data types together.

student <- list(
  "Rayhan",
  23,
  TRUE
)

print(student)
class(student)
```

```r
#Matrix is ​​used to store similar data in Row × Column format.

mat <- matrix(c(1, 2, 3, 4, 5, 6),
              nrow = 2,
              ncol = 3)

print(mat)
```

```r
#Array is the Similar to the Matrix but can have more dimensions than 2D.

arr <- array(1:12,
dim = c(2,3,2)
)

print(arr)
```

```r
#Data Frame is A table-like data structure where the data types of the columns can be different.

students <- data.frame(
  Name = c('Rayhan', 'Chowdhury', 'Nir'),
  Age = c(22,33,55),
  Passed = c(TRUE, FALSE,TRUE),
  Home = c('Faridpur','Dhaka','CTG')
)

print(students)
```

Vector     → Same Type

List       → Different Type

Matrix     → 2D Same Type

Array      → Multi-Dimensional

Data Frame → Table
