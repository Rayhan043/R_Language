#R Plotting


```r
#Plot is show relation with two veriables
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 6, 8, 10)

plot(x, y)
```

```r
#line Chart
plot(x, y, type = "l")
```

```r
# bar Chart

marks <- c(80, 60, 90)

barplot(marks, names.arg = c("Math", "English", "ICT"))
```

```r
# Pic Chart
data <- c(40, 30, 20, 10)

pie(data, labels = c("A", "B", "C", "D"))
```

#ggplot 2

```r
install.packages("ggplot2")
library(ggplot2)
```

```r
# Scatter Plot
data <- data.frame(
  x = c(1,2,3,4,5),
  y = c(2,4,6,8,10)
)

ggplot(data, aes(x=x, y=y)) +
  geom_point()
```

```r
# Line Chart
ggplot(data, aes(x=x,y=y) ) +
  geom_line()
```

```r
# Bar Chart
data2 <- data.frame(
  sub = c('Math', 'English', 'Bangla'),
  marks = c(80,70,90)
)

ggplot(data2, aes(x = sub, y = marks)) +
  geom_bar(stat = "identity")

```

```r
#  Histogram (Distribution)

data3 <- data.frame(
  catagory = c('A','B','C','D'),
  value = c(30,40,50,60)
)

ggplot(data3, aes(x= "", y = value, fill = catagory)) +
  geom_bar(stat = 'identity')+
  coord_polar('y')
```

```r
# Histogram
data3 <- data.frame(
  marks = c(55,60,65,70,75,80,85,90,95)
)

ggplot(data3, aes(x=marks)) +
  geom_histogram(binwidth=10)
```

#Memory Tricks

| Plot      | Function                  |
| --------- | ------------------------- |
| Scatter   | geom_point()              |
| Line      | geom_line()               |
| Bar       | geom_bar(stat="identity") |
| Histogram | geom_histogram()          |
| Pie       | coord_polar()             |


```r
memory = data.frame(
  Plot = c('Scatter','Line','Bar','Histogram','Pie'),
  # cat("----------\n")
  Function = c('geom_point()','geom_line()','geom_bar(stat="identity")','geom_histogram()','geom_polar()')
)
print(memory)
```

```r
library(tibble)

memory <- tibble(
  Plot = c('Scatter','Line','Bar','Histogram','Pie'),
  Function = c('geom_point()','geom_line()','geom_bar(stat="identity")','geom_histogram()','coord_polar()')
)

View(memory)
```
