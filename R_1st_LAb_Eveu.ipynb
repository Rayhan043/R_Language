#Pie chart

```r
x <- c(30,20,25,15,10)
labels <- c('A','B','C','D','E')
pie(x, labels, col=rainbow(5),main="Sample pie chart")
```

#Box plot

```r
boxplot(Petal.Length ~ Species, data=iris,
        col=c("red","green","blue"),
        main="Petal Length by Species",
        xlab="Species", ylab="Petal Length")
```

```r
boxplot(Sepal.Width ~ Species, data=iris,
        col=rainbow(3),
        main="Sepal Width Distribution by Species",
        xlab="Species", ylab="Sepal Width")
```

#Line Plot

```r
plot(
  mtcars$hp	, mtcars$mpg,
  type = "l",
  main= 'Line Plot of Horsepower vs Mileage',
  xlab='HP',
  ylab='Mil'
)
```

```r
# Scatter plot
plot(
  mtcars$wt , mtcars$qsec,
  pch = 19,
  col = "darkgreen",
  main = "Scatter Plot of Weight vs Quarter Mile Time",
  xlab = "wt",
  ylab = "qsce"
)
```

```r
# PirChart
cyl_count <- table(mtcars$cyl)
pie(
cyl_count,
    main = "Pie Chart of Cylinder Count",
    xlab = "Cylinder (cyl)",
    ylab = "Frequency"
)
```

```r
# BoxPlot
boxplot(
  mpg ~ cyl,
  data = mtcars,
  main = "Boxplot of MPG by sylinder",
  xlab = "Num od cyl",
  ylab = "Num of MGP"
)
```

```r
# Bar plot

gear_count <- table(mtcars$gear)

barplot(gear_count,
        main = "Bar Plot of Gear Count",
        xlab = "Number of Gears (gear)",
        ylab = "Frequency")
```
